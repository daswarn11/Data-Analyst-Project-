```python
# Essential libraries
import pandas as pd          # For data manipulation and analysis
import numpy as np           # For numerical operations

# Visualization libraries
import matplotlib.pyplot as plt  # For plotting graphs
import seaborn as sns            # For advanced visualizations

# Configure some visualization settings
%matplotlib inline   # Display plots inline in Jupyter Notebook
sns.set_style("whitegrid")   # Set a nice style for seaborn plots
```

```python
# Let's start by loading the data from the CSV file and doing a quick exploratory data analysis (EDA).

# Load the dataset
file_path = 'marketing_campaign.csv'
df = pd.read_csv(file_path)

# Display the first few rows to understand the structure of the dataset
print (df.head())
```

```python
# General information about the dataset
df.info()

# Checking for missing values
df.isnull().sum()
 
```

```python
# Get descriptive statistics for numerical columns
desc_stats = df.describe(include='all')

desc_stats
```

```python
# Filling missing values in 'Income' column with the median
df['Income'].fillna(df['Income'].median(), inplace=True)
```

```python
# Distribution of Income
plt.figure(figsize=(8, 5))
sns.histplot(df['Income'], bins=30, kde=True)
plt.title('Income Distribution')
plt.show()

# Boxplot for Wines Spending
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['MntWines'])
plt.title('Spending on Wines Boxplot')
plt.show()
```

```python

# Select only numerical columns
numeric_df = df.select_dtypes(include=['float64', 'int64'])

# Compute the correlation matrix
corr_matrix = numeric_df.corr()

# Create a mask for the upper triangle
mask = np.triu(np.ones_like(corr_matrix, dtype=bool))

# Set up the matplotlib figure
plt.figure(figsize=(14, 10))

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr_matrix, mask=mask, annot=True, cmap='coolwarm', fmt=".2f", 
            linewidths=1, linecolor='white', cbar_kws={"shrink": .8}, annot_kws={"size": 10})

# Set the title
plt.title('Correlation Matrix for Numerical Features', fontsize=16)
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.yticks(rotation=0)  # Ensure y-axis labels are horizontal
plt.show()
```

```python
# Countplot of campaign response
plt.figure(figsize=(6, 4))
sns.countplot(x='Response', data=df)
plt.title('Response to the Last Campaign')
plt.show()

# Percentage of customers who responded to the campaign
response_rate = df['Response'].mean() * 100
print(f'Response Rate: {response_rate:.2f}%')
```

```python
# Separate customers who responded vs those who did not
responders = df[df['Response'] == 1]
non_responders = df[df['Response'] == 0]

# Check the size of both groups
print(f"Number of Responders: {len(responders)}")
print(f"Number of Non-Responders: {len(non_responders)}")

# Basic statistics of key features for responders
responders_stats = responders.describe()

# Basic statistics of key features for non-responders
non_responders_stats = non_responders.describe()

responders_stats, non_responders_stats
```

```python
# Analyze response rates by income, education, and other key features
response_by_income = df.groupby(['Income'])['Response'].mean() * 100
response_by_education = df.groupby(['Education'])['Response'].mean() * 100
response_by_marital_status = df.groupby(['Marital_Status'])['Response'].mean() * 100

print("Response Rate by Income:")
print(response_by_income)

print("\nResponse Rate by Education:")
print(response_by_education)

print("\nResponse Rate by Marital Status:")
print(response_by_marital_status)
```

```python
# Visualize response rates by income, education, and marital status
plt.figure(figsize=(10, 6))
sns.boxplot(x='Income', y='Response', data=df)
plt.title('Response by Income')
plt.show()

plt.figure(figsize=(8, 6))
sns.barplot(x='Education', y='Response', data=df)
plt.title('Response by Education Level')
plt.show()

plt.figure(figsize=(8, 6))
sns.barplot(x='Marital_Status', y='Response', data=df)
plt.title('Response by Marital Status')
plt.show()
```

```python
# Average spending by response status
spending_cols = ['MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts', 'MntSweetProducts', 'MntGoldProds']
response_spending = df.groupby('Response')[spending_cols].mean()

response_spending.plot(kind='bar', figsize=(10, 6))
plt.title('Average Spending by Response Status')
plt.ylabel('Average Spending')
plt.xticks(rotation=0)
plt.show()
```

```python
# Boxplot for Recency by Response
plt.figure(figsize=(8, 6))
sns.boxplot(x='Response', y='Recency', data=df)
plt.title('Recency Distribution by Response Status')
plt.show()
```

```python
# Convert 'Dt_Customer' to datetime format if not already
df['Dt_Customer'] = pd.to_datetime(df['Dt_Customer'])

# Analyze the response rate based on when the customer joined
df['Year_Customer'] = df['Dt_Customer'].dt.year
response_by_year = df.groupby('Year_Customer')['Response'].mean() * 100

plt.figure(figsize=(10, 6))
sns.lineplot(x=response_by_year.index, y=response_by_year.values)
plt.title('Response Rate by Year of Customer Joining')
plt.ylabel('Response Rate (%)')
plt.show()
```
